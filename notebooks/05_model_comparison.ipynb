# 05 - Model Comparison

Compare available model metrics from `reports/metrics`.

```python
from pathlib import Path
import json
import pandas as pd

METRICS_DIR = Path('..') / 'reports' / 'metrics'

def load_accuracy(path, model_name):
    if not path.exists():
        return {'model': model_name, 'accuracy': None, 'source': str(path)}
    metrics = json.loads(path.read_text(encoding='utf-8'))
    return {'model': model_name, 'accuracy': metrics.get('accuracy'), 'source': str(path)}

comparison = pd.DataFrame([
    load_accuracy(METRICS_DIR / 'metrics.json', 'LightGBM'),
    load_accuracy(METRICS_DIR / 'random_forest_metrics.json', 'Random Forest'),
])
comparison.sort_values('accuracy', ascending=False, na_position='last')
```

```python
ax = comparison.dropna(subset=['accuracy']).plot(
    kind='bar',
    x='model',
    y='accuracy',
    legend=False,
    color=['#2463eb', '#237a57'],
    ylim=(0, 1),
)
ax.set_title('Model accuracy comparison')
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy')
```

## Selection rule

Use the model with the highest validation accuracy after checking class-level precision and recall. Keep LightGBM as the production default unless a comparison model is both better and simpler to operate.
