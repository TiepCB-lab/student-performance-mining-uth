# 01 - Data Understanding

Inspect the raw student performance dataset, schema, missing values, and target distribution.

```python
from pathlib import Path
import pandas as pd

DATA_PATH = Path('..') / 'data' / 'raw' / 'Student_Performance.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
summary = pd.DataFrame({
    'dtype': df.dtypes.astype(str),
    'missing': df.isna().sum(),
    'unique': df.nunique(),
})
summary
```

```python
df['final_grade'].value_counts().sort_index().rename('rows').to_frame()
```

## Notes

- Target column: `final_grade`.
- `student_id` is an identifier and should not be used as a feature.
